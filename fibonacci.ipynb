```python
import numpy as np
import time 
```

```python
# Code that was given to me and I had to decipher
def fibby(num):
    if type(num) != int and num < 0:
        print('Error, please input a non-negative integer')
    else:
        if num == 0:
            return 0
        elif num == 1:
            return 1
        else:
            x1 = fibby(num - 1)
            x2 = fibby(num - 2)
            return x1 + x2
```

```python
# My take on the Algorythm
def fibby2(num):
    gr = (1+np.sqrt(5))/2
    if num == 0:
        return 0
    elif num == 1:
        return 1
    else:
        x = 1
        for i in range(num-2):
            x = round(x * gr) 
        return x
    
    
```

```python
# Running Tests
def check_fibby(x1,x2):
    if x1 == x2:
        print('True')
    else: 
        print('False', x1,x2)

def run(number):
    for i in range(number):
        print(i)
        st1 = time.time()
        x1 = fibby(i)
        et1 = time.time()
        print(x1, ': fibby1 took', round(et1 - st1), 'seconds') 
        st2 = time.time()
        x2 = fibby2(i)
        et2 = time.time()
        print(x2, ': fibby2 took', round(et2 - st2), 'seconds') 
        check_fibby(x1,x2)
```

```python
run(100)
```
